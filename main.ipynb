## 3 Limpieza y Normalizacion de los datos

En esta etapa se realiza la limpieza del dataset para prepararlo para el analisis

* Primero se trantan los valores faltantes utilizando un unico metodo ( en este usare la media )
* Luego se transforma la columnas "FECHA" a datatime para poder trabajar correctamente
* Por ultimo , se normalizan los nombre y apellidos corrigiendo las mayusculas

```python
df.dropna(inplace=True)
```

```python
df["Fecha"] = pd.to_datetime(df["Fecha"], errors="coerce")
```

```python
df["Nombre_Cliente"] = df["Nombre_Cliente"].str.title()
df["Apellido_Cliente"] = df["Apellido_Cliente"].str.title()
```

## 4 Obtencion de Estaditica Descriptiva

Luego del Proceso de Limpiza , se vuelve a calcular las estaditicas descriptivas para analizar como se han modificado los datos
ademas se realiza un agrupamiento por dos columnas para obtener estadisticas especificas por grupo

```python
df.describe()
```

```python
df.groupby(["Cantidad", "Precio_Unitario"]).mean(numeric_only=True)
```

## 5 Visualizacion de datos con Matplotlib

Se utiliza un Grafico de Barras para visualizar el total de ventas por prodcuto esto permite analizar que productos
generaron mas ingresos en el periodo

```python
import matplotlib.pyplot as plt

ventas_por_productos = df.groupby("Producto")["Total_Venta"].sum()
plt.figure(figsize=(10,6))
ventas_por_productos.plot(kind="bar", color="lightgreen")
plt.title("Ventas Totales por Productos")
plt.xlabel("Productos")
plt.ylabel("Total de ventas")
plt.xticks(rotation=90, ha="right")
plt.tight_layout()
plt.show()
```

## 6 Exportancion a Archivo SQlite

en esta seccion se exportan los datos procesados a una base de datos SQlite

```python
import sqlite3

conn = sqlite3.connect("Ventas.db")

df.to_sql("ventas", conn, if_exists="replace", index=False)

consulta_sql = pd.read_sql_query("SELECT * FROM ventas LIMIT 5", conn)

print(consulta_sql)

conn.close()
```
